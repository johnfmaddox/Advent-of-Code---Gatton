**--- Part One ---**

```python
## Select input data case
## Note: this assumes a plain textfile named `input_{case}` is located in the same folder as this notebook
case = "example" # <- input_example
case = "jfm"
# case = "gatton"

# verbose output
verbose = False

import numpy as np

# Read input
with open(f'input_{case}','r') as f:
    input_lines = f.readlines()

import numpy as np

class Grid():
    def __init__(self):
        self.to_int = {' ':0, '.':1, '#':2}
        self.to_sym = {0:' ', 1:'.', 2:'#',4:'>',5:'v',6:'<',7:'^'}
        self.n_row = len(input_lines[:-2])
        self.n_col = max(len(i) for i in input_lines[:-2])-1
        self.grid = np.zeros((self.n_row,self.n_col))
        for i,row in enumerate(input_lines[:-2]):
            for j,val in enumerate(row[:-1]):
                self.grid[i,j] = self.to_int[val]
        
        self.show_grid = self.grid.copy()
            
        self.directions = []
        tmp = ''
        for i in input_lines[-1].strip():
            if i in ['R','L']:
                self.directions.append([int(tmp),i])
                tmp=''
            else:
                tmp += i
        self.pos = np.array([0,np.where(self.grid[0,:]==1)[0][0]])
        self.facing = 0
        self.row_len = len(self.grid[0])
        self.col_len = len(self.grid)
    
    def update_show_grid(self):
        self.show_grid[self.pos[0],self.pos[1]] = 4+self.facing

    def execute(self,count=None,verbose=False):
        if verbose: print('executing')
        for i,dir in enumerate(self.directions):
            if count is not None:
                if i >= count:
                    return None
            if verbose: print(dir)
            for i in range(dir[0]):
                self.move_once(verbose=verbose)
            self.turn(dir[1])

    def turn(self,dir):
        if dir == 'R':
            self.facing += 1
        elif dir == 'L':
            self.facing -= 1
        self.facing = (self.facing+4)%4
        self.update_show_grid()
    
    def show(self):
        # show_grid = self.grid.copy()
        # show_grid[self.pos[0],self.pos[1]] = self.facing+4
        for i,row in enumerate(self.show_grid):
            print(''.join([self.to_sym[i] for i in row])+str(i))

    def move_once(self,verbose=False):
        f = self.facing
        pos = self.pos
        grid = self.grid
        if f==0: # right
            if verbose: print('moving right')
            search_array = np.append(grid[pos[0],pos[1]:],grid[pos[0],:pos[1]-1])
            try:
                open_idx = np.where(search_array[1:] == 1)[0][0]+1
            except IndexError:
                open_idx=0
            try:
                wall_idx = np.where(search_array[1:] == 2)[0][0]+1
            except IndexError:
                wall_idx = open_idx + 1
            if open_idx < wall_idx:
                self.pos += [0,open_idx]
                if self.pos[1]>self.n_col-1:
                    self.pos[1] = self.pos[1]-self.n_col
        if f == 2: # left
            if verbose: print('moving left')
            search_array = np.flip(np.append(grid[pos[0],pos[1]+1:],grid[pos[0],:pos[1]+1]))
            try:
                open_idx = np.where(search_array[1:] == 1)[0][0]+1
            except IndexError:
                open_idx=0            
            try:
                wall_idx = np.where(search_array[1:] == 2)[0][0]+1
            except IndexError:
                wall_idx = open_idx + 1
            if open_idx < wall_idx:
                self.pos -= [0,open_idx]
                if self.pos[1]<0:
                    self.pos[1] = self.n_col+self.pos[1]
        if f == 1: # down
            if verbose: print('moving down')
            search_array = np.append(grid[pos[0]:,pos[1]],grid[:pos[0]-1,pos[1]])
            try:
                open_idx = np.where(search_array[1:] == 1)[0][0]+1
            except IndexError:
                open_idx=0
            try:
                wall_idx = np.where(search_array[1:] == 2)[0][0]+1
            except IndexError:
                wall_idx = open_idx + 1
            if open_idx < wall_idx:
                self.pos += [open_idx,0]
                if self.pos[0]>self.n_row-1:
                    self.pos[0] = self.pos[0]-self.n_row 
        if f == 3: # up
            if verbose: print('moving up')
            search_array = np.flip(np.append(grid[pos[0]+1:,pos[1]],grid[:pos[0]+1,pos[1]]))
            if verbose: print(search_array)
            try:
                open_idx = np.where(search_array[1:] == 1)[0][0]+1
            except:
                open_idx=0
            try:
                wall_idx = np.where(search_array[1:] == 2)[0][0]+1
            except IndexError:
                wall_idx = open_idx + 1
            if verbose: print(f'{open_idx=}; {wall_idx=}')
            if open_idx < wall_idx:
                self.pos -= [open_idx,0]
                if verbose: print(f'pos={self.pos}')
                if self.pos[0]<0:
                    self.pos[0] = self.n_row+self.pos[0]
                if verbose: print(f'pos={self.pos}')
        self.update_show_grid()
    
    def password(self):
        return 1000*(self.pos[0]+1) + 4*(self.pos[1]+1) + self.facing

grid = Grid()
```

```python
grid = Grid()

grid.execute(count=None,verbose=False)
# print(grid.pos)
# grid.show()
grid.password()
```
