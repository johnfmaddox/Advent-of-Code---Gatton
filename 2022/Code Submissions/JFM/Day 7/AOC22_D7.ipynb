```python
## Select input data case
## Note: this assumes a plain textfile named `input_{case}` is located in the same folder as this notebook
case = "example" # <- input_example
case = "jfm"
case = "gatton"

# verbose output
verbose = False

# Read input into a list of strings by line
with open(f'input_{case}','r') as f:
    input_lines = f.readlines()

input_lines

def get_dir(pwd):
    result = tree
    for i in pwd:
        result = result[i]
    return result

tree =  {}

dirs = []
pwd = []
curdir = tree
for line in input_lines:
    line = line.strip().split()
    if line[0]=='$':
        if line[1] == 'cd':
            arg = line[2]
            if arg == '..':
                dirsize = curdir['size']
                pwd.pop()
                curdir = get_dir(pwd)
                curdir['size'] += dirsize
            else:
                curdir[arg] = {'basename':arg, 'files':[], 'dirs':[], 'size':0}
                pwd.append(arg)
                dirs.append(pwd.copy())
                curdir = get_dir(pwd)
    else:
        if line[0] == 'dir':
            arg = line[1]
            curdir['dirs'].append(arg)
        else:
            curdir['size'] += int(line[0])
            curdir['files'].append(line[1])

# Walk back to root directory to include size of nested directories if needed
while pwd != ['/']:
    dirsize = curdir['size']
    pwd.pop()
    curdir = get_dir(pwd)
    curdir['size'] += dirsize

total_size = 0
for path in dirs:
    dir = get_dir(path)
    size = dir['size']
    if size <= 100000:
        total_size += size
print(f'Total: {total_size}')
```
