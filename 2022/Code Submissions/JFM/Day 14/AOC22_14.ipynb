```python
## Select input data case
## Note: this assumes a plain textfile named `input_{case}` is located in the same folder as this notebook
case = "example" # <- input_example
case = "jfm"
case = "gatton"

# verbose output
verbose = False

import numpy as np
from itertools import cycle

# Read input
with open(f'input_{case}','r') as f:
    input_lines = f.readlines()

import numpy as np

conv = {0:'.', 1:'O', 2:'#'}

class Grid():
    def __init__(self, part=1):
        self.grains = 0
        self.off_map = False
        self.rock_paths = []
        self.mini = 0
        self.minj = 500
        self.maxi = 0
        self.maxj = 500

        for line in input_lines:
            split_line = line.strip().split('->')
            coords = []
            for coords_str in split_line:
                coords.append(coords_str.split(','))
                i = int(coords[-1][1])
                j = int(coords[-1][0])
                # print(f'{i=}, {j=}')
                self.mini = min([self.mini, i])
                self.maxi = max([self.maxi, i])
                self.minj = min([self.minj, j])
                self.maxj = max([self.maxj, j])
            self.rock_paths.append([[int(coord[1]),int(coord[0])] for coord in coords])
        # print(f'{self.maxi}, {self.maxj}')

        self.part = part
        if part == 2:
            self.maxi += 2
            self.maxj += self.maxi+2
        self.grid = np.zeros([self.maxi+1, self.maxj+1])
        if part == 2:
            self.grid[-1,:] = 2
        for path in self.rock_paths:
            for i in range(len(path)-1):
                i1,j1 = path[i]
                i2,j2 = path[i+1]
                if i1>i2:
                    tmp = i2
                    i2 = i1
                    i1 =tmp
                if j1>j2:
                    tmp = j2
                    j2 = j1
                    j1 = tmp
                # print(f'#: {i1},{j1} -> {i2},{j2}')
                self.grid[i1:i2+1,j1:j2+1] = 2

    def in_map(self, pos):
        i,j = pos
        if (
            i < self.mini or i > self.maxi or
            j < self.minj or j > self.maxj
        ):
            return False
        return True

    def drop_sand(self, start=(0,500)):
        pos = np.array(tuple(start))
        finished = False

        while not finished:
            try:
                if self.grid[tuple(pos+(1,0))] == 0:
                    pos += (1,0)
                elif self.grid[tuple(pos+(1,-1))] == 0:
                    pos += (1,-1)
                elif self.grid[tuple(pos+(1,1))] == 0:
                    pos += (1,1)
                else:
                    self.grid[tuple(pos)] = 1
                    finished = True
            except IndexError:
                self.off_map = True
                finished = True
        if not self.off_map:
            self.grains += 1

    def drop_sand_continuous(self):
        if self.part == 1:
            while not self.off_map:
                self.drop_sand()
        elif self.part == 2:
            while not self.grid[0,500]:
                self.drop_sand()

    def show(self):
        for line in self.grid[self.mini:self.maxi+1, self.minj:self.maxj+1]:
            print(f'{"".join([conv[val] for val in line])}')
```

```python
grid = Grid()
grid.drop_sand_continuous()
grid.show()
print(grid.grains)
```
